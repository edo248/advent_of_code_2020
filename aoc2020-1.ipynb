```python
import io

def parse(input_data):
    indata = [line.strip() for line in io.StringIO(input_data)]
    return list(map(int, indata))

data = parse(input_data)
```

```python
def find(numbers, limit):    
    possibles=[limit-numbers[0]]
    for number in numbers[1:]:
        if number in possibles:
            return (number*(limit-number))            
        else:
            possibles.append(limit-number)

find(numbers=data, limit=2020)
```

```python
for number in data:
    found = find(data, limit=2020-number)    
    if found:
        result = found * number
        print(f"Found {result} for {number}")
```
